```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import fetch_california_housing

df = fetch_california_housing(as_frame=True).frame
df.head(10)
```

```python
print(df.describe())
```

```python
print(df.columns)
```

```python
def fun(X):
  plt.hist(df[X],bins=20,color=['b'],alpha=0.5)
  plt.xlabel(X)
  plt.ylabel("Frequency")
  plt.title("Histogram of "+X)
  plt.show()

for i in df.columns:
  print(fun(i))

```

```python
sns.boxplot(x=df['MedInc'],color='b')
plt.xlabel('MedInc')
plt.ylabel('frequency')
plt.show()
```

```python
#Display outliers for the selected attribute
# Calculate the first (Q1) and third quartiles (Q3)
Q1 = df['MedInc'].quantile(0.25)
Q3 = df['MedInc'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Determine the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['MedInc'] < lower_bound) | (df['MedInc'] > upper_bound)]
print("Outliers:\n", outliers)
```
