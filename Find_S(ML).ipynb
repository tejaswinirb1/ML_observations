```python
import pandas as pd

def find_s_algorithm(data):

    # Extract features and labels
    features = data.iloc[:, :-1].values  # All columns except the last (features)
    labels = data.iloc[:, -1].values  # Last column (target concept)

    # Initialize the most specific hypothesis
    num_attributes = features.shape[1]
    hypothesis = ["∅"] * num_attributes

    # Iterate through training examples
    for i in range(len(features)):
        if labels[i] == "Yes":  # Consider only positive examples
            if hypothesis == ["∅"] * num_attributes:
                hypothesis = list(features[i])  # Initialize with first positive example
            else:
                for j in range(num_attributes):
                    if hypothesis[j] != features[i][j]:  # Generalize where different
                        hypothesis[j] = "?"

    return hypothesis

def fill_missing_values(data):

    # Apply Find-S to get the most specific hypothesis
    hypothesis = find_s_algorithm(data)

    # Replace missing values in the dataset based on the hypothesis
    for i, column in enumerate(data.columns[:-1]):  # Ignore the last column (class label)
        if hypothesis[i] != "?":  # Only replace missing values if hypothesis has a fixed value
            data[column] = data[column].fillna(hypothesis[i])

    # Fill missing values in the target concept column based on the hypothesis
    target_column = data.columns[-1]  # Last column (target concept)
    for i in range(len(data)):
        if pd.isna(data.loc[i, target_column]):  # If target column is NaN
            row_features = data.iloc[i, :-1].values  # Get feature values of the row
            if all(h == "?" or h == row_features[j] for j, h in enumerate(hypothesis)):
                data.loc[i, target_column] = "Yes"  # Fill with "Yes" if it matches the hypothesis
            else:
                data.loc[i, target_column] = "No"  # Otherwise, assume "No"

    return data

# ✅ Read the CSV file first before calling functions
csv_file = "/content/data_with_headers.csv"
data = pd.read_csv(csv_file)  # Read CSV into DataFrame

# ✅ Pass DataFrame instead of file path
print("Learned Hypothesis:", find_s_algorithm(data))

# Fill missing values using the learned hypothesis
updated_data = fill_missing_values(data)

# Save the cleaned data back to CSV (optional)
updated_data.to_csv("updated_data.csv", index=False)

# Display the updated dataset
print("Updated Dataset:")
print(updated_data)
```
